```python
#imported libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Problem 1: Iris Dataset


```python
# Load datasets
iris = pd.read_csv('iris.txt', sep='\t')
ames = pd.read_csv('ames_housing.txt', sep='\t')
titanic = pd.read_csv('titanic.txt', sep='\t')
batting = pd.read_csv('batting.csv', sep=',')
```

```python
#1. Data frame for iri DataFrame
#Display it with head()
iris .head(8)
```

```python
#2. Display some descriptive statistics for each of the columns in the iris DataFrame.
iris.describe()
```

```python
#3. Apply Grouping to all the three species
iris_means = iris.groupby('species').mean()
iris_means
```

```python
#4. Use a bar chart to graphically represent the information calculated above.
# Creating the lists of colors and column names
iris_colors = ['red', 'green', 'blue']
iris_columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

# Creating the figure with 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

for axs,column, title in zip(axs.flatten(), iris_means.columns, iris_columns):
    axs.bar(iris_means.index, iris_means[column], color=iris_colors, edgecolor='black')
    axs.set_title(f"Mean {title}")
plt.tight_layout()
plt.show()
# axes = axes.flatten()

# for i, ax in enumerate(axes):
#     ax.bar(iris_means.index, iris_means.iloc[:, i], color=iris_colors, edgecolor='black')
#     ax.set_title(f"Mean {iris_columns[i]}")

```

## Problem 2: Ames Housing - Neighborhoods

```python
# Display the first 8 rows of the ames DataFrame
ames.head(8)
```

```python
#2. Display dataframe using the columns attribute of a DataFrame to view the names of the columns it contains.
print(ames.columns.values)

#ames.columns.values
```

```python
#3. calculate the average sale price for houses in for each neighborhood.
ames_nbhd = ames[['Neighborhood',  'SalePrice']].groupby('Neighborhood').mean().sort_values('SalePrice')
ames_nbhd.head(5)
```

```python
# Creating a horizontal bar chart for mean sale price by neighborhood
plt.figure(figsize=(6, 10))

# Plotting the horizontal bar chart
plt.barh(ames_nbhd.index, ames_nbhd['SalePrice'], color='skyblue', edgecolor='black') 

# Setting labels and title
plt.xlabel('Mean Sale Price')
plt.ylabel('Neighborhood')
plt.title('Mean Sale Price by Neighborhood')

# Displaying the chart
plt.show()
```

## Problem 3: Ames Housing – House Styles

## Problem 6: Batting Data – Home Runs

```python
# print the names of the columns in this DataFrame
print(batting.columns.values)
```

```python
# Calculate the most home runs in a single season
total_hr = batting[['teamID', 'yearID', 'HR']].groupby(['teamID', 'yearID'], as_index=False).sum().sort_values('HR', ascending=False)

# Display the first 10 rows of total_hr DataFrame
total_hr.head(10)
```

```python
# Calculate the highest average number of home runs per season
avg_hr = total_hr[['teamID', 'HR']].groupby('teamID').mean().sort_values(by='HR', ascending=False)
avg_hr.head(10)
```

## Problem 7: Batting Data – Batting Averages

```python
# Create batting_avg DataFrame
batting_avg = batting[['teamID', 'yearID', 'H', 'AB']].groupby(['teamID', 'yearID']).sum().reset_index()
batting_avg['BA'] = batting_avg['H'] / batting_avg['AB']
batting_avg.sort_values('BA', ascending=False).head(10)
```

## Problem 8: Batting Data – Cardinals vs. Cubs

```python
# Calculate batting statistics for the Cardinals
stl_batting = batting.loc[(batting['teamID'] == 'SLN') & (batting['yearID'] >= 1900), ['yearID', 'H', 'AB', 'HR']].groupby('yearID').sum()
stl_batting['BA'] = stl_batting['H'] / stl_batting['AB']

# Calculate batting statistics for the Cubs
chi_batting = batting.loc[(batting['teamID'] == 'CHN') & (batting['yearID'] >= 1900), ['yearID', 'H', 'AB', 'HR']].groupby('yearID').sum()
chi_batting['BA'] = chi_batting['H'] / chi_batting['AB']

# Create the figure with two line plots
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Plotting batting averages
axs[0].plot(stl_batting.index, stl_batting['BA'], label='Cardinals', color='red')
axs[0].plot(chi_batting.index, chi_batting['BA'], label='Cubs', color='blue')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Batting Average')
axs[0].set_title('Batting Average By Year')
axs[0].legend()

# Plotting home runs
axs[1].plot(stl_batting.index, stl_batting['HR'], label='Cardinals', color='red')
axs[1].plot(chi_batting.index, chi_batting['HR'], label='Cubs', color='blue')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Home Runs')
axs[1].set_title('Home Runs by Year')
axs[1].legend()

# Display the figure
plt.show()
```

```python
# Calculate the proportion of years with higher batting average for the Cardinals
higher_batting_avg = np.mean(stl_batting['BA'] > chi_batting['BA'])
print(f"Proportion of years with higher batting average for the Cardinals: {round(higher_batting_avg, 4)}")
```

```python
# Calculate the proportion of years with more home runs for the Cardinals
more_home_runs = np.mean(stl_batting['HR'] > chi_batting['HR'])
print(f"Proportion of years with more home runs for the Cardinals: {round(more_home_runs, 4)}")
```
