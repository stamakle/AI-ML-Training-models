## Problem 1: Sample Mean and Variance


```python
#importing libraries
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Problem 1: Sample Mean and Variance
# Step 1: Create array x
x = np.array([10, 16, 26, 12, 17, 22, 14, 12, 21, 16])

# Step 2: Store length of array in n
n = len(x)

# Step 3: Calculate sample mean (using formula)
mean = np.sum(x) / n

# Step 4: Calculate the differences between each value and the mean
diff = x - mean

# Step 5: Calculate sample variance (using formula)
var = np.sum(diff**2) / (n - 1)

# Step 6: Print results
print(f"Sample Mean:     {mean:.4f}")
print(f"Sample Variance: {var:.4f}")

# Now, using numpy's built-in functions
mean_np = np.mean(x)
var_np = np.var(x, ddof=1)  # Setting ddof=1 for sample variance

print(f"\nUsing NumPy Functions:")
print(f"Sample Mean:     {mean_np:.4f}")
print(f"Sample Variance: {var_np:.4f}")
```

```python
#calculate the second mean and variance
mean_np = np.mean(x)
var_np = np.var(x, ddof=1)

#Print the results
print("Sample Mean (using np.mean()):", mean_np)
print("Sample Variance (using np.var()):", var_np)

```

## Problem 2: Scoring a Regression Model

```python
# Problem 2: Scoring a Regression Model
# Define the function to calculate SSE
def find_sse(true_y, pred_y):
    return np.sum((true_y - pred_y) ** 2)

# Data provided in the problem
true_y = np.array([22.1, 17.9, 16.5, 14.3, 19.8, 23.7, 22.0, 18.4, 25.7, 19.2])
pred_1 = np.array([21.4, 16.7, 17.9, 12.1, 22.1, 25.1, 21.7, 19.3, 23.4, 19.9])
pred_2 = np.array([20.7, 18.1, 16.9, 13.6, 21.9, 24.8, 20.3, 21.1, 24.8, 18.4])

# Calculate SSE for both models
sse_1 = find_sse(true_y, pred_1)
sse_2 = find_sse(true_y, pred_2)

# Print the results
print(f"Model 1 SSE: {sse_1:.2f}")
print(f"Model 2 SSE: {sse_2:.2f}")
```

## Problem 3: Scoring a Classification Model

```python
# Problem 3: Scoring a Classification Model
# Define the function to calculate accuracy
def find_accuracy(true_y, pred_y):
    return np.mean(true_y == pred_y)

# Medical diagnosis data
true_diag = np.array(['P', 'P', 'N', 'N', 'P', 'N', 'N', 'N', 'P', 'N', 'N', 'N', 'N', 'P', 'P', 'N', 'N', 'N', 'N', 'N'])
pred_diag = np.array(['N', 'P', 'N', 'P', 'P', 'N', 'P', 'N', 'P', 'N', 'N', 'N', 'P', 'P', 'P', 'N', 'N', 'N', 'P', 'N'])

# Calculate and print accuracy for medical diagnosis
accuracy_diag = find_accuracy(true_diag, pred_diag)
print(f"Model Accuracy: {accuracy_diag:.4f}")

# Image classification data
true_labels = np.array(['dog', 'dog', 'cat', 'dog', 'cat', 'cat', 'cat', 'dog', 'cat', 'cat', 'dog', 'cat', 
                        'cat', 'dog', 'dog', 'dog', 'dog', 'cat', 'cat', 'cat', 'dog', 'dog', 'cat', 'cat'])
pred_labels = np.array(['dog', 'dog', 'cat', 'dog', 'cat', 'dog', 'cat', 'dog', 'cat', 'cat', 'dog', 'cat', 
                        'cat', 'dog', 'cat', 'dog', 'dog', 'cat', 'dog', 'cat', 'dog', 'dog', 'cat', 'cat'])

# Calculate and print accuracy for image classification
accuracy_labels = find_accuracy(true_labels, pred_labels)
print(f"Model Accuracy: {accuracy_labels:.4f}")
```

## Problem 4: Classification Report

```python
# Problem 4: Classification Report
def classification_report(true_y, pred_y):
    classes = np.unique(true_y)
    negative_class, positive_class = classes[0], classes[1]
    
    accuracy = find_accuracy(true_y, pred_y)
    
    # Calculate TP, FP, TN, FN
    TP = np.sum((pred_y == positive_class) & (true_y == positive_class))
    FP = np.sum((pred_y == positive_class) & (true_y == negative_class))
    TN = np.sum((pred_y == negative_class) & (true_y == negative_class))
    FN = np.sum((pred_y == negative_class) & (true_y == positive_class))
    
    # Calculate precision and recall
    pos_precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    pos_recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    neg_precision = TN / (TN + FN) if (TN + FN) > 0 else 0
    neg_recall = TN / (TN + FP) if (TN + FP) > 0 else 0
    
    # Print results
    print(f"Positive Class:      {positive_class}")
    print(f"Negative Class:      {negative_class}")
    print(f"\nAccuracy:            {accuracy:.4f}")
    print(f"Positive Precision:  {pos_precision:.4f}")
    print(f"Positive Recall:     {pos_recall:.4f}")
    print(f"Negative Precision:  {neg_precision:.4f}")
    print(f"Negative Recall:     {neg_recall:.4f}")

# Applying the function to medical diagnosis data
print("\nClassification Report for Medical Diagnosis:")
classification_report(true_diag, pred_diag)

# Applying the function to image classification data
print("\nClassification Report for Image Classification:")
classification_report(true_labels, pred_labels)
```

## Problem 5: Transformation of Random Variables

```python
np.random.seed(1)
X = np.random.normal(0,0.4, 25000)
Y = np.exp(X)

mean_X = np.mean(X)
std_X = np.std(X, ddof=1)
mean_Y = np.mean(Y)
std_Y = np.std(Y, ddof=1)

print('Sample Mean of X: \t',round(mean_X, 4) )
print('Sample Std Dev of X: \t',round(std_X, 4) )
print('Sample Mean of Y: \t',round(mean_Y, 4) )
print('Sample Std Dev of Y: \t',round(std_Y, 4) )
```

```python
# Create histograms of X and Y
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Histogram of X values
axs[0].hist(X, bins=30, edgecolor='black', color='blue')
axs[0].set_title("Histogram of X Values")

# Histogram of Y values
axs[1].hist(Y, bins=30, edgecolor='black', color='green')
axs[1].set_title("Histogram of Y Values")

# Display the figure with histograms
plt.show()
```

```python
# Calculate the probabilities
prob_0_5 = np.mean(Y < 0.5)
prob_1 = np.mean(Y < 1.0)
prob_2 = np.mean(Y < 2.0)

# Print the probabilities
print("Probability that Y is less than 0.5:", format(prob_0_5, '.4f'))
print("Probability that Y is less than 1.0:", format(prob_1, '.4f'))
print("Probability that Y is less than 2.0:", format(prob_2, '.4f'))
```

## Problem 6: Stochastic Linear Relationships

```python
# Set the random seed
np.random.seed(1)

# Sample values for x_vals and errors
x_vals = np.random.normal(10, 2, 200)
errors = np.random.normal(0, 1.2, 200)

# Calculate y_vals according to the equation Y = 5.1 + 0.9*X + errors
y_vals = 5.1 + 0.9 * x_vals + errors

# Create a scatter plot of x_vals and y_vals
plt.figure(figsize=(8, 6))
plt.scatter(x_vals, y_vals, s=60, alpha=0.8, edgecolors='black', color='blue')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Display the scatter plot
plt.show()
```

```python
# Calculate the correlation between x_vals and y_vals
diff_x = x_vals - np.mean(x_vals)
diff_y = y_vals - np.mean(y_vals)
numerator = np.sum(diff_x * diff_y)
denominator = np.sqrt(np.sum(diff_x**2) * np.sum(diff_y**2))
correlation = numerator / denominator

# Print the correlation
print("Correlation between X and Y:", format(correlation, '.4f'))
```

## Problem 7: Relationship between Life Expectancy and Per Capita GDP

```python
import pandas as pd

# Read the data into a DataFrame
df = pd.read_csv('gapminder_data.txt', sep='\t')

# Extract the required columns
country = df.country.values
year = df.year.values
continent = df.continent.values
population = df.population.values
life_exp = df.life_exp.values
pcgdp = df.gdp_per_cap.values
gini = df.gini.values

# Clean up memory
df = None

# Create a list of continents and colors
continent_list = ['africa', 'americas', 'asia', 'europe']
color_list = ['red', 'blue', 'green', 'orange']

# Set figure size
plt.figure(figsize=(8, 6))

# Loop over continents and create scatter plots
for i, cont in enumerate(continent_list):
    # Boolean masking for current continent and year 2018
    sel = (continent == cont) & (year == 2018)
    
    # Scatter plot with selected data
    plt.scatter(np.log(pcgdp[sel]), life_exp[sel], s=100, alpha=0.7, edgecolors='black', color=color_list[i], label=cont.title())

# Set labels and title
plt.xlabel('Natural Log of Per Capita GDP')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs Per Capita GDP (2018)')

# Add legend
plt.legend()

# Display the plot
plt.show()

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Loop over continents and create subplots
for i, cont in enumerate(continent_list):
    # Boolean masking for current continent and year 2018
    sel = (continent == cont) & (year == 2018)
    
    # Create subplot
    ax = axs[i // 2, i % 2]
    
    # Scatter plot with selected data
    ax.scatter(np.log(pcgdp[sel]), life_exp[sel], s=100, alpha=0.7, edgecolors='black', color=color_list[i])
    
    # Set subplot limits, labels, and title
    ax.set_xlim([6, 12])
    ax.set_ylim([45, 90])
    ax.set_xlabel('Natural Log of Per Capita GDP')
    ax.set_ylabel('Life Expectancy')
    ax.set_title(cont.title())

# Adjust subplot spacing
plt.tight_layout()

# Display the plot
plt.show()
```
